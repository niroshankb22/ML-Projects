```python
import pandas as pd
import  matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import recall_score,f1_score,precision_score,roc_auc_score,accuracy_score
```

```python
data = pd.read_csv('restaurant_customer_satisfaction.csv')
```

```python
le = LabelEncoder()

data['Gender'] = le.fit_transform(data['Gender'])
data['VisitFrequency'] = le.fit_transform(data['VisitFrequency'])
data['PreferredCuisine'] = le.fit_transform(data['PreferredCuisine'])
data['TimeOfVisit'] = le.fit_transform(data['TimeOfVisit'])
data['DiningOccasion'] = le.fit_transform(data['DiningOccasion'])
data['MealType'] = le.fit_transform(data['MealType'])
```

```python
data.head()
```

```python
data.info()
```

```python
X = data.drop('HighSatisfaction',axis=1)
Y =data['HighSatisfaction']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
data
```

# LogisticRegression

```python
LR = LogisticRegression(class_weight='balance')
```

```python
LR.fit(X_train,Y_train)
```

```python
Y_pred = LR.predict(X_test)
```

```python
threshold = 0.4
y_pred_custom = (Y_pred >= threshold).astype(int)
```

```python
precision_score(Y_test,Y_pred)
```

```python
recall_score(Y_test,Y_pred)
```

```python
f1_score(Y_test,Y_pred)
```

```python
roc_auc_score(Y_test,Y_pred)
```

```python
accuracy_score(Y_test,Y_pred)
```

# RandomForestClassifier

```python
X = data.drop('HighSatisfaction',axis=1)
Y =data['HighSatisfaction']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
RFC = RandomForestClassifier(class_weight='balanced')
```

```python
RFC.fit(X_train,Y_train)
```

```python
Y_pred1 = RFC.predict(X_test)
```

```python
f1_score(Y_test,Y_pred1)
```

```python
recall_score(Y_test,Y_pred1)
```

```python
precision_score(Y_test,Y_pred1)
```

```python
roc_auc_score(Y_test,Y_pred1)
```

```python
accuracy_score(Y_test,Y_pred1)
```

# SVC

```python
from sklearn.svm import SVC
```
