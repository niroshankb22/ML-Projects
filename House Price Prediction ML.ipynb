```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('house_data.csv')
```

```python
data
```

```python
data.info()
```

```python
data=data[['price','waterfront','sqft_living','condition','bedrooms','bathrooms','floors','grade']]
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
SSC = StandardScaler()
```

```python
data[['sqft_living','condition','bedrooms','bathrooms','floors','grade']]=SSC.fit_transform(data[['sqft_living','condition','bedrooms','bathrooms','floors','grade']])
```

```python
data
```

```python
X = data.drop('waterfront',axis=1)
y = data['waterfront']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
```

```python
KNC = KNeighborsClassifier(n_neighbors=5,metric='l2')
```

```python
KNC.fit(X_train,y_train)
```

```python
y_pred =KNC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
